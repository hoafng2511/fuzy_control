```python
# 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 2. Import
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# 3. Data generator (tự resize ảnh về 60x60)
datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    "/content/drive/MyDrive/data_mono/",
    target_size=(60, 60),
    batch_size=32,
    class_mode="categorical",
    subset="training"
)

val_generator = datagen.flow_from_directory(
    "/content/drive/MyDrive/data_mono/",
    target_size=(60, 60),
    batch_size=32,
    class_mode="categorical",
    subset="validation"
)

# In ra class để biết tên người ứng với số
print("Class indices:", train_generator.class_indices)

# 4. Model CNN
model = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(60, 60, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation="relu"),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.5),
    Dense(train_generator.num_classes, activation="softmax")
])

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()

# 5. Train
model.fit(train_generator, validation_data=val_generator, epochs=10)

# 6. Save model
model.save("/content/drive/MyDrive/face_model.h5")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
from google.colab import files

# Load model đã train
model = load_model("/content/drive/MyDrive/face_model.h5")

# Lấy class_indices từ train_generator khi train model
classes = {'NguyenVanHoa': 0, 'NguyenDuongNhatHuy': 1, 'VoHoangKhang': 2}

# Danh sách thông tin theo đúng thứ tự class_id
info_list = [
    "Nguyễn Văn Hòa - MSSV 31241021473 - Sinh ngày 25/11/2006 - Lớp ICA",
    "Lê Nhật Huy - MSSV 31231026121 - Sinh ngày 5/11/2005 - Lớp Robot",
    "Võ Hoàng Khang - MSSV 31241021883 - Sinh ngày 9/11/2006 - Lớp ICA"
]

# Upload ảnh bất kỳ
uploaded = files.upload()
for fn in uploaded.keys():
    img = cv2.imread(fn)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img_rgb, (60, 60)) / 255.0
    img_ready = np.expand_dims(img_resized, axis=0)

    # Dự đoán
    pred = model.predict(img_ready)
    class_id = np.argmax(pred)

    # Lấy tên + info
    class_name = list(classes.keys())[class_id]
    class_info = info_list[class_id]

    # In kết quả
    print("Ảnh vừa upload:", fn)
    print("Dự đoán:", class_name)
    print("👉 Thông tin:", class_info)

    # Hiển thị ảnh
    plt.imshow(img_rgb)
    plt.title(f"{class_name}\n{class_info}")
    plt.axis("off")
    plt.show()

```
