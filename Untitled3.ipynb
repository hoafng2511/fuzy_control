```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# ===== 1. Khai báo biến mờ =====
water_level = ctrl.Antecedent(np.arange(0, 101, 1), 'water_level')  # cm
inflow_rate = ctrl.Antecedent(np.arange(0, 51, 1), 'inflow_rate')   # L/min
pump_power = ctrl.Consequent(np.arange(0, 101, 1), 'pump_power')    # %

# ===== 2. Định nghĩa membership functions =====
water_level['low'] = fuzz.trimf(water_level.universe, [0, 0, 50])
water_level['medium'] = fuzz.trimf(water_level.universe, [30, 50, 70])
water_level['high'] = fuzz.trimf(water_level.universe, [50, 100, 100])

inflow_rate['low'] = fuzz.trimf(inflow_rate.universe, [0, 0, 20])
inflow_rate['medium'] = fuzz.trimf(inflow_rate.universe, [10, 25, 40])
inflow_rate['high'] = fuzz.trimf(inflow_rate.universe, [30, 50, 50])

pump_power['off'] = fuzz.trimf(pump_power.universe, [0, 0, 0])
pump_power['low'] = fuzz.trimf(pump_power.universe, [0, 25, 50])
pump_power['medium'] = fuzz.trimf(pump_power.universe, [25, 50, 75])
pump_power['high'] = fuzz.trimf(pump_power.universe, [50, 100, 100])

# ===== 3. Định nghĩa đủ 9 luật fuzzy =====
rule1 = ctrl.Rule(water_level['low'] & inflow_rate['low'], pump_power['high'])
rule2 = ctrl.Rule(water_level['low'] & inflow_rate['medium'], pump_power['medium'])
rule3 = ctrl.Rule(water_level['low'] & inflow_rate['high'], pump_power['low'])

rule4 = ctrl.Rule(water_level['medium'] & inflow_rate['low'], pump_power['high'])
rule5 = ctrl.Rule(water_level['medium'] & inflow_rate['medium'], pump_power['medium'])
rule6 = ctrl.Rule(water_level['medium'] & inflow_rate['high'], pump_power['low'])

rule7 = ctrl.Rule(water_level['high'] & inflow_rate['low'], pump_power['medium'])
rule8 = ctrl.Rule(water_level['high'] & inflow_rate['medium'], pump_power['low'])
rule9 = ctrl.Rule(water_level['high'] & inflow_rate['high'], pump_power['off'])

# ===== 4. Tạo hệ thống điều khiển =====
pump_ctrl = ctrl.ControlSystem([rule1, rule2, rule3,
                                 rule4, rule5, rule6,
                                 rule7, rule8, rule9])

pump_sim = ctrl.ControlSystemSimulation(pump_ctrl)

# ===== 5. Thử nhập dữ liệu =====
pump_sim.input['water_level'] = 90  # cm
pump_sim.input['inflow_rate'] = 45   # L/min

# Tính toán
pump_sim.compute()

print("Kết quả công suất bơm:", pump_sim.output['pump_power'])

pump_power.view(sim=pump_sim)
```
